# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

```python
res=requests.get('http://www.imdb.com/chart/top')
soup=BeautifulSoup(res.text,"html.parser")
```

```python
movies=soup.select('td.titleColumn')
#movies
movieInfo = movies[0].get_text()
print(movieInfo)
movie = (' '.join(movieInfo.split()).replace('.', ''))
print(movie)
```

```python
#print(soup.select('td.posterColumn span[name=ir]'))
ratings=[rating_span.attrs.get('data-value')for rating_span in soup.select('td.posterColumn span[name=ir]')]
#print(ratings)
```

```python
movies_details=[]
for index in range(0,100):
    movieInfo=movies[index].get_text()
    arrangMovieInfo=(' '.join(movieInfo.split()).replace('.',''))
    movieName=arrangMovieInfo[len(str(index))+1:-7]
    yearOfRelease=re.search('\((.*?)\)', movieInfo).group(1)
    movies_detail={"Name": movieName,
            "Rating": ratings[index],
            "Year of release": yearOfRelease
            }
    movies_details.append(movies_detail)
#movies_details
```

```python
df = pd.DataFrame(movies_details)
df
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape: 


# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
req=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup=BeautifulSoup(req.text,"html.parser")
```

```python
#result=soup.find_all('tr',attrs={'class':'rankings-block__banner'})

#for table in soup.find_all('table'):
#    print(table.get('class'))

df=pd.DataFrame(columns=['Position','Team Name','Team Code','Matches','Points','Rating'])
table = soup.find('table', class_='table')
count=1
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    if(columns!=[] and count<11):
        pos=columns[0].text.strip()
        team=columns[1].text.strip()
        team_name=team.split('\n',1)[0]
        team_code=team.split('\n',1)[1]
        matches=columns[2].text.strip()
        points=columns[3].text.strip()
        rating=columns[4].text.strip()
        df=df.append({'Position':pos,'Team Name':team_name,'Team Code':team_code,'Matches':matches,'Points':points,'Rating':rating},ignore_index=True)
        count+=1
            
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :

i) Restaurant name

ii) Cuisine

iii) Location 

iv) Ratings

v) Image URL

```python
req=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup=BeautifulSoup(req.text,"html.parser")
```

```python
result=soup.find_all('div',attrs={'class':'restnt-card restaurant'})
#result
```

```python
result[0]
```

```python
print(result[0].find('div',class_='restnt-rating rating-4 hide').text)
print(result[0].find('img')['data-src'])
print(result[0].find('a').text)
#print(result[0].find('span').find_all('a')[0].text)

foods=[]
for i in result[0].find('span').find_all('a'):
    foods.append(i.text)
    
print(','.join(foods))

loc_info=[]
for j in result[0].find('div',class_='restnt-loc ellipsis').find_all('a'):
    loc_info.append(j.text)
    
print(','.join(loc_info))
```

```python
restaurant_info=[]

for r in range(len(result)):
    res_name=result[r].find('a').text.strip()
    foods=[]
    for i in result[r].find('span').find_all('a'):
        foods.append(i.text)
    cuisine=','.join(foods)
    loc_info=[]
    for j in result[r].find('div',class_='restnt-loc ellipsis').find_all('a'):
        loc_info.append(j.text)
    location=','.join(loc_info)
    rating=result[r].find('div',class_='restnt-rating rating-4').text
    image_url=result[r].find('img')['data-src']
    restaurant={'Restaurant name':res_name,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Image URL':image_url}
    restaurant_info.append(restaurant)
    
restaurant_info
```

```python
df=pd.DataFrame(restaurant_info)
df
```

```python
data=requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
soup=BeautifulSoup(data.text,"html.parser")
```

```python
for row in table.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    if(columns!=[]):
        c1=columns[0].text
        c2=columns[1].text
        c3=columns[2].text
        c4=columns[3].text
print(c1)
print(c2)
print(c3)
print(c4)
```
